```python
import numpy as np
import pandas as pd

# Load the DataSet
df = pd.read_csv("C:/Users/Abdullah Umer/Desktop/Data Sentinel Internship/Task4/Shopping_Trends.csv")

# Quick Preview of Dataset
print("✅ Dataset Preview:")
print(df.head())
print("\n Columns:", df.columns.tolist())
print("\n Basic Info:")
print(df.info())

# 1. Average Purchase Amount by Gender
avg_purchase_gender = df.groupby("Gender")["Purchase Amount (USD)"].mean().reset_index()
print("\n✅ Average Purchase Amount by Gender:")
print(avg_purchase_gender)

# 2. Total Purchase Amount by Category
total_purchase_category = df.groupby("Category")["Purchase Amount (USD)"].sum().reset_index()
print("\n✅ Total Purchase Amount by Category:")
print(total_purchase_category)

# 3. Average Age by Location
avg_age_location = df.groupby("Location")["Age"].mean().reset_index()
print("\n✅ Average Age by Location:")
print(avg_age_location)

# 4. Customer Count by Subscription Status
customer_count_subscription = df.groupby("Subscription Status")["Customer ID"].count().reset_index()
customer_count_subscription.rename(columns={"Customer ID": "Customer Count"}, inplace=True)
print("\n✅ Customer Count by Subscription Status:")
print(customer_count_subscription)

# 5. Multiple Aggregations: Gender-wise Summary
gender_summary = df.groupby("Gender").agg(
    Avg_Age=("Age", "mean"),
    Total_Spend=("Purchase Amount (USD)", "sum"),
    Avg_Spend=("Purchase Amount (USD)", "mean"),
    Count=("Customer ID", "count")
).reset_index()

print("\n✅ Gender-wise Summary (Aggregated):")
print(gender_summary)
```
